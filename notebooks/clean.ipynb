Voy a llevar a cabo un estudio y actualizado de campos utilizando regex. 

```python
import re # Importo paquete regex

# Defino una función que  busque los caracteres clave en la cadena de entrada
def busca_swim(cadena):

    patron = r'swim' # Defino un patrón de busqueda

    resultado = re.search(patron, cadena, re.IGNORECASE)

    if resultado: # Si el resultado no está vacío quiere decir que lo ha encontrado
        return 'Swimming'
    else:
        return cadena # En caso de que no la encuentre se devuelve cadena original
```

```python
res = busca_swim('SWImidning in da pool') # Compruebo que funcione
print(res)
```

```python
res = busca_swim('cadena sin palabra')
print(res)
```

```python
# Ahora voy a hacer esta función exactamente igual para los otros casos de diving, fishing y surfing
```

```python
# Defino una función que  busque los caracteres clave en la cadena de entrada
def busca_dive(cadena):

    cadena = cadena.lower()

    patron = r'div' # Defino un patrón de busqueda

    resultado = re.search(patron, cadena)

    if resultado: # Si el resultado no está vacío quiere decir que lo ha encontrado
        return 'Diving'
    else:
        return cadena # En caso de que no la encuentre se devuelve cadena original
```

```python
# Defino una función que  busque los caracteres clave en la cadena de entrada
def busca_fish(cadena):

    cadena = cadena.lower()

    patron = r'fish' # Defino un patrón de busqueda

    resultado = re.search(patron, cadena)

    if resultado: # Si el resultado no está vacío quiere decir que lo ha encontrado
        return 'Fishing'
    else:
        return cadena # En caso de que no la encuentre se devuelve cadena original
```

```python
# Defino una función que  busque los caracteres clave en la cadena de entrada
def busca_surf(cadena):

    cadena = cadena.lower()

    patron = r'surf' # Defino un patrón de busqueda

    resultado = re.search(patron, cadena)

    if resultado: # Si el resultado no está vacío quiere decir que lo ha encontrado
        return 'Surfing'
    else:
        return cadena # En caso de que no la encuentre se devuelve cadena original
```

```python
# Defino una función que  busque los caracteres clave en la cadena de entrada
def busca_over(cadena):

    cadena = cadena.lower()

    patron = r'over' # Defino un patrón de busqueda

    resultado = re.search(patron, cadena)

    if resultado: # Si el resultado no está vacío quiere decir que lo ha encontrado
        return 'Overboard or capsized'
    else:
        return cadena # En caso de que no la encuentre se devuelve cadena original
```

```python
# Defino una función que  busque los caracteres clave en la cadena de entrada
def busca_board(cadena): # Las actividades que sean de tipo board las voy a meter en surfing también

    cadena = cadena.lower()

    patron = r'board' # Defino un patrón de busqueda

    resultado = re.search(patron, cadena)

    if resultado: # Si el resultado no está vacío quiere decir que lo ha encontrado
        return 'Surfing'
    else:
        return cadena # En caso de que no la encuentre se devuelve cadena original
```

Procedo a aplicar las 4 fórmulas que he definido a la columna `activity` con el formato apply

```python
dc.info()
```

```python
dc.activity = dc.activity.astype(str)
```

```python
# Procedo a aplicar las 4 fórmulas a la columna activity del dataframe
dc.activity = dc.activity.apply(busca_swim)
dc.activity = dc.activity.apply(busca_dive)
dc.activity = dc.activity.apply(busca_surf)
dc.activity = dc.activity.apply(busca_fish)
dc.activity = dc.activity.apply(busca_over) # Llamo primero a esta función para quitarme los overboard antes de los deportes de tabla
dc.activity = dc.activity.apply(busca_board)
```
